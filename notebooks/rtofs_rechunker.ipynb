```python
from rechunker import rechunk
```

```python
target_chunks = {
    'salinity': {'time': 1096, 'depth': 40, 'Y': 57, 'X': 106},
    'temperature': {'time': 1096, 'depth': 40, 'Y': 57, 'X': 106},
    'u': {'time': 1096, 'depth': 40, 'Y': 57, 'X': 106},
    'v': {'time': 1096, 'depth': 40, 'Y': 57, 'X': 106},
    'time': None, # don't rechunk this array
    'lon': None,
    'lat': None,
    'X': None,
    'Y': None
}
max_mem = '1.25GB'
```

```python
target_store = 'rtofs_rechunked.zarr'
temp_store = 'rtofs_rechunked-tmp.zarr'

# need to remove the existing stores or it won't work
# !rm -rf rtofs_rechunked.zarr rtofs_rechunked-tmp.zarr
array_plan = rechunk(ds, target_chunks, max_mem, target_store, temp_store=temp_store)
array_plan
```

```python
from dask.diagnostics import ProgressBar

with ProgressBar():
    array_plan.execute()
```
